## 4. Data Preparation

```python
cars_data
```

```python
X = cars_data[["hp",	"drat",	"wt",	"qsec",	"vs",	"am",	"gear",	"carb","cyl"	,"disp"]]
X
```

```python
y=cars_data["mpg"]
y
```

### there is nothing to prepare because all features are numeric

## 5.Model Building

```python
 linear_model = LinearRegression()
```

## 6.Model Training

```python
linear_model.fit(X,y)
```

```python
linear_model.intercept_
```

```python
linear_model.coef_
```

## 7.Model Testing

```python
y_pred = linear_model.predict(X)
y_pred
```

## 8.Model Evaluation

```python
error = y - y_pred
error
```

### Let's now perform Test 5: Homoscedasticity Test

```python
sns.scatterplot(data=cars_data,x = "qsec",y = error)
plt.show()
```

### Test 5: Homoscedasticity Test Also Failed

## Test 6: Zero Residual Mean Test

```python
sns.scatterplot(x = cars_data["mpg"],y = y_pred)
plt.show()
```

## 9.Model Deployment
